```python
def find_mrml_files(directory):
    # Use os.path.join to ensure the path is constructed correctly for the OS
    path = os.path.join(directory, '**', '*.mrml')

    # Use glob.glob with recursive set to True to find all .mrml files
    files = glob.glob(path, recursive=True)

    return files

def load_mrml_scene(filePath):
    slicer.mrmlScene.Clear()
    slicer.util.loadScene(filePath)
    
    
    
    
def get_names_of_volumes():    
    """
    first output is name of study 0 and second of study 1
    """
    # Get all volume nodes
    volumeNodes = slicer.mrmlScene.GetNodesByClass('vtkMRMLScalarVolumeNode')
    names=[]
    # Iterate over all volume nodes and print their names
    for i in range(volumeNodes.GetNumberOfItems()):
        volumeNode = volumeNodes.GetItemAsObject(i)
        # print(volumeNode.GetName())    
        names.append(volumeNode.GetName())
    names=list(filter(lambda el: "Standard" in el, names))    
    names=list(filter(lambda el: "transf" not in el, names))    
    sudy_0=list(filter(lambda el: "sudy_0" in el, names))[0]    
    sudy_1=list(filter(lambda el: "sudy_1" in el, names))[0]    
    return sudy_0,sudy_1
        
def apply_transform(volume_names, transformation_name="LinearTransform"):
    """
    apply transformation to the study 0 SUV image 
    defaoult transformaion is LinearTransform - that was prepared manually by fidicial registration
    """
    ##### applying transformation to the node
    # Get the transform node
    transformNode = slicer.util.getNode(transformation_name)

    # Get the volume node
    volumeNode = slicer.util.getNode(volume_names[0])

    # Apply the transform to the volume
    volumeNode.SetAndObserveTransformNodeID(transformNode.GetID())
        
        
def get_auto_fiducial_transform():
     # Get the markup nodes
    fromMarkupNode = slicer.util.getNode('From')
    toMarkupNode = slicer.util.getNode('To')

    # Create a new linear transform node
    linTransfNode = slicer.vtkMRMLLinearTransformNode()
    linTransfNode.SetName('lin_transf')
    slicer.mrmlScene.AddNode(linTransfNode)

    # Assuming that both markups have the same number of fiducials
    numFiducials = fromMarkupNode.GetNumberOfFiducials()

    # Create matrices to store fiducial points
    fromPoints = vtk.vtkPoints()
    toPoints = vtk.vtkPoints()

    # Get fiducial points from both markups
    for i in range(numFiducials):
        coord = [0, 0, 0]
        fromMarkupNode.GetNthFiducialPosition(i, coord)
        fromPoints.InsertNextPoint(coord)
        toMarkupNode.GetNthFiducialPosition(i, coord)
        toPoints.InsertNextPoint(coord)

    # Create landmark transform
    landmarkTransform = vtk.vtkLandmarkTransform()
    landmarkTransform.SetSourceLandmarks(fromPoints)
    landmarkTransform.SetTargetLandmarks(toPoints)
    landmarkTransform.SetModeToRigidBody()  # or landmarkTransform.SetModeToAffine()
    # landmarkTransform.SetModeToAffine()  # or landmarkTransform.SetModeToAffine()
    landmarkTransform.Update()

    # Set the transform matrix to the linear transform node
    linTransfNode.SetMatrixTransformToParent(landmarkTransform.GetMatrix())

    
    
def create_ants_registration_transform(volume_names):
    # Get the volume nodes
    fixedVolumeNode = slicer.util.getNode(volume_names[1])
    movingVolumeNode = slicer.util.getNode(volume_names[0])

    # Create a new transform node
    transformNode = slicer.vtkMRMLLinearTransformNode()
    transformNode.SetName('ants_transform')
    slicer.mrmlScene.AddNode(transformNode)

    # Set up the parameters for the registration
    parameters = {}
    parameters['fixedVolume'] = fixedVolumeNode.GetID()
    parameters['movingVolume'] = movingVolumeNode.GetID()
    parameters['outputTransform'] = transformNode.GetID()
    parameters['initializeTransformMode'] = 'useCenterOfHeadAlign'
    parameters['transformType'] = 'Rigid'  # Add this line

    # Run the registration using ANTs
    slicer.cli.runSync(slicer.modules.antsregistration, None, parameters)    
    
    
def create_general_registration_transform(volume_names):
    ####using python create transformation called """gen_transf""" that would register volume node """pat_10_sudy_0_SUV""" to volume node """pat_10_sudy_1_2023-06-15_Standardized_Uptake_Value_body_weight""" using General Registration (BRAINS) module

    # Get the volume nodes
    fixedVolumeNode = slicer.util.getNode(volume_names[1])
    movingVolumeNode = slicer.util.getNode(volume_names[0])

    # Create a new transform node
    transformNode = slicer.vtkMRMLLinearTransformNode()
    transformNode.SetName('general_transform')
    slicer.mrmlScene.AddNode(transformNode)

    # Set up the parameters for the registration
    parameters = {}
    parameters['fixedVolume'] = fixedVolumeNode.GetID()
    parameters['movingVolume'] = movingVolumeNode.GetID()
    parameters['outputTransform'] = transformNode.GetID()
    parameters['initializeTransformMode'] = 'useCenterOfHeadAlign'
    parameters['transformType'] = 'Rigid'  # Add this line

    # Run the registration
    slicer.cli.runSync(slicer.modules.brainsfit, None, parameters)

def create_elastix_registration_transform(volume_names):
    # Get the volume nodes
    fixedVolumeNode = slicer.util.getNode(volume_names[1])
    movingVolumeNode = slicer.util.getNode(volume_names[0])

    # Get the Elastix logic

    # Create a new transform node
    transformNode = slicer.vtkMRMLLinearTransformNode()
    transformNode.SetName('elastix_transform')
    slicer.mrmlScene.AddNode(transformNode)

    #taken from https://discourse.slicer.org/t/accessing-slicerelastix-from-within-python-module/9410/4
    # elastixLogic = Elastix.ElastixLogic()
    # parameterFilenames = elastixLogic.getRegistrationPresets()[0][Elastix.RegistrationPresets_ParameterFilenames]
    # elastixLogic.registerVolumes(
    #                                 fixedVolumeNode, movingVolumeNode,
    #                                 # parameterFilenames = parameterFilenames,
    #                                 outputTransformNode = transformNode
    #                                 )
    
    
    # Set up the parameters for the registration
    parameters = {}
    parameters['fixedVolume'] = fixedVolumeNode.GetID()
    parameters['movingVolume'] = movingVolumeNode.GetID()
    # parameters['outputTransform'] = transformNode.GetID()
    elastixLogic.RegisterVolumes(parameters)
    parameters['initializeTransformMode'] = 'useCenterOfHeadAlign'
    parameters['transformType'] = 'Rigid'  # Add this line

    # Run the registration
    slicer.cli.runSync(slicer.modules.elastix, None, parameters)    
    
    
    
    
    
    
    
    
def get_fiducial_points(markupNode):
    """reading coordinates of markups"""
    # Get the markup node
    numPoints = markupNode.GetNumberOfFiducials()
    coords=[]
    for i in range(numPoints):
        coord = [0, 0, 0]
        markupNode.GetNthControlPointPositionWorld(i, coord)
        coords.append(coord)
    return coords 

    
def get_fiducial_points_from_name(name):
    """reading coordinates of markups"""
    # Get the markup node
    markupNode = slicer.util.getNode(name)    
    return get_fiducial_points(markupNode)
    
    
def calculate_control_point_distance(transform_name):
    ###### reading coordinates of markups
    # Get the markup node
    markupNode = slicer.util.getNode("From")
    markupNode.SetAndObserveTransformNodeID(None)#at the begining we do not want it registered
    if(transform_name!="None"):
        # print(f"apply transform {transform_name}")
        # markupNode = slicer.util.getNode("To")
        transformNode = slicer.util.getNode(transform_name)
        markupNode.SetAndObserveTransformNodeID(transformNode.GetID())
        coords_To= get_fiducial_points(markupNode)
        
    else:
        coords_To= get_fiducial_points_from_name('From')
    
    
    coords_From= get_fiducial_points_from_name('To')
    
    res=0.0
    # print(transform_name)
    
    for i in range(len(coords_From)):
        point_a=coords_From[i]
        point_b=coords_To[i]
        # print(f"p {point_a}   {point_b}")
        loc_res=0.0
        for axis in [0,1,2]:
            loc_res=loc_res+((point_a[axis]-point_b[axis])**2)
        res=res+np.sqrt(loc_res)
    return res   
            

    

    
```
